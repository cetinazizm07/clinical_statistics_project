```python
import pandas as pd
import numpy as np

# 1. Define the file path
file_path = "/Users/azizcetin/clinical_statistics_project/data/raw/stat_clinical_data.xlsx" 

# 2. Load the dataset (Sheet ST12, starting from row index 3)
df = pd.read_excel(file_path, sheet_name='ST12', na_values=["NA"], header=3)

# Clean column names (removes accidental spaces or newlines from Excel headers)
df.columns = df.columns.str.strip()

print("==================================================")
print(f"FULL DATASET SUMMARY")
print(f"Total Patients: {df.shape[0]} | Total Columns: {df.shape[1]}")
print("==================================================\n")

# 3. Loop through EVERY single column in the dataframe
for col in df.columns:
    print(f"{'='*50}")
    print(f"COLUMN: [{col}]")
    
    # Calculate missing values for the specific column
    missing_count = df[col].isnull().sum()
    missing_pct = (missing_count / len(df)) * 100
    print(f"Missing Values: {missing_count} ({missing_pct:.2f}%)")
    print(f"{'-'*50}")
    
    # If the column contains numbers (e.g., Age, Tumor size, Days from diagnosis)
    if pd.api.types.is_numeric_dtype(df[col]):
        print("Type: NUMERIC - Summary Statistics:")
        display(df[col].describe().round(2))
        
    # If the column contains text/categories (e.g., PAM50, Treatments, Receptors)
    else:
        print("Type: CATEGORICAL - Distribution:")
        # Display counts for every unique value, including NaNs
        display(df[col].value_counts(dropna=False))
        
    print("\n")
```
